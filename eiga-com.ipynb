```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import time
import re
import math


# from google.colab import drive
# drive.mount('/content/drive')
# %cd "/content/drive/MyDrive/Colab Notebooks/ゼミ/映画/"
```

```python
def scrape_review(movie_id):
  url = 'https://eiga.com/movie/' + str(movie_id) + '/review'
  page = requests.get(url)
  html = BeautifulSoup(page.text, 'html.parser')
  title = html.select('.title-link a')[0].text
  reviews = html.select('.txt-block p')
  total_number = html.select('.result-number')
  total_num = int(re.findall("(?<=\全).+?(?=\件)", total_number[0].text)[0])

  reviews_list = [reviews[i].text for i in range(len(reviews))]

  for i in range(1, math.ceil(total_num / 20)):
    time.sleep(0.5)
    url = 'https://eiga.com/movie/' + str(movie_id) + '/review/all/' + str(i+1)
    page = requests.get(url)
    html = BeautifulSoup(page.text, 'html.parser')
    reviews = html.select('.txt-block p')
    reviews_list += [reviews[j].text for j in range(len(reviews))]

  return title, reviews_list
```

```python
movie_ids = [91554, 91634, 95716, 94998, 91393]
# 1104, 1077, 191, 167, 183
```

```python
for id in movie_ids:
  title, reviews = scrape_review(id)
  print(title)
  print('reviewの数: ', len(reviews))
  series = pd.Series(data=reviews, name='review')
  title = '_'.join(title.split())
  series.to_csv(title + '_reviews.csv', index=False)
```
