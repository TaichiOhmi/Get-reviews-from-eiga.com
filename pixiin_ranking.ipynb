```python
# pixiin.com から 2000年から2021年までの興行収入ランキング(日本)のランキングを取得
all_ranking = {}
for i in range(2000, 2022):
  url = 'https://pixiin.com/ranking-japan-boxoffice' + str(i)
  page = requests.get(url)
  html = BeautifulSoup(page.text, 'html.parser')
  lines = html.find_all("div", class_="tb3n", limit=50)
  ranking_list = []
  for j in range(len(lines)):
    if lines[j].select_one('.t_sr'):
      lines[j].select_one('.t_sr').decompose()
    ranking_list.append((lines[j].text))
  all_ranking[str(i)] = ranking_list
  time.sleep(1)
```

```python
# google

headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/62.0.3202.94 Safari/537.36"}
# for y in range(2021, 2010, -1):
#   year = str(y)
year = '2021'

for i in range(2, len(all_ranking[year])):

  # URL に &が含まれる場合、エンコード
  if '&' in all_ranking[year][i]:
    url_name = all_ranking[year][i].replace('&', '%26')
  else:
    url_name = all_ranking[year][i]

  # URL
  url = 'https://www.google.com/search?q=' + urllib.parse.quote(url_name + ' 映画.com eiga.com ' + year) + '&hl=ja'

  # print(url)
  page = requests.get(url, headers=headers)
  time.sleep(3)
  html = BeautifulSoup(page.text, 'html.parser')
  time.sleep(3)
  movie_url = html.find("div", class_="yuRUbf")
  time.sleep(3)
  link = movie_url.find('a').get('href')
  movie_id = re.search("(?<=\/)\d+?(?=\/)", link).group()
  all_ranking[year][i] = [all_ranking[year][i], int(movie_id)]
  print(i, all_ranking[year][i])
  time.sleep(5)
```

```python
# yahoo
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/62.0.3202.94 Safari/537.36"}

for i in range(2009, 2022):
  year = str(i)

  for i in range(len(all_ranking[year])):
      if '&' in all_ranking[year][i]:
        url_name = all_ranking[year][i].replace('&', '%26')
      else:
        url_name = all_ranking[year][i]
      url = 'https://search.yahoo.co.jp/search?p=' + url_name + '+映画.com+eiga.com+' + year + '&ei=UTF-8'
      # print(url)
      page = requests.get(url, headers = headers)
      time.sleep(3)
      html = BeautifulSoup(page.text, 'html.parser')
      time.sleep(1)
      # web = html.select_one("#web")
      # time.sleep(2)
      link = html.find('a', class_='sw-Card__titleInner').get('href')
      time.sleep(1)
      movie_id = re.search("(?<=\/)\d+?(?=\/)", link).group()
      all_ranking[year][i] = [all_ranking[year][i], int(movie_id)]
      print(i, all_ranking[year][i])
      time.sleep(2)
```
